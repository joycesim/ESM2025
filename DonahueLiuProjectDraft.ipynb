$$\frac{ζ_{1,j}^{m+1}}{T}=-c\frac{ζ_{1,j}^{m}-ζ_{1,j-1}^{m}}{h}$$
Where j is space index, h is space step

Define $ζ_1=ζ_1^* λ^m e^{iκjh}$ as a boundary condition onto the hyperbolic formula, allowing us to simplify to a diffusive problem

Where $λ = e^{iωT}$ and is called stability parameter, or amplification factor

Courant number being $q=\frac{cT}{h}$.
* To be stable, cT must be less than h, or distance traveled per time step is less than grid size h.
